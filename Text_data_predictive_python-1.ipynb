1.Basic feature extraction using text data
    -No of words
    -No of characters
    -Average word length
    -No of stopwords
    -No of special charaters
    -No of numerics
    -No of uppercase words

```python
import pandas as pd
```

```python
train = pd.read_csv("D:\\Datasets\\train_E6oV3lV.csv")
train.head()
```

```python
#No. of words
train['word_count'] = train['tweet'].apply(lambda x:len(str(x).split(" ")))
#train['word_count']
```

```python
train[['tweet','word_count']].head()
```

```python
#No. of characters
```

```python
train['char_count']=train['tweet'].str.len()
train[['tweet','char_count']].head()
```

```python
#Avg words
```

```python
def avg_word(sen):
    words = sen.split()
    return (sum(len(word) for word in words))/len(words)
train['avg_word'] = train['tweet'].apply(lambda x: avg_word(x))
train[['tweet','avg_word']].head()
```

```python
#No of stopwords
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('English')
stop
```

```python
train['tweet'].apply(lambda x: len([x for x in x.split() if x in stop]))
```

```python
#Number of special characters
```

```python
train['hashtags'] = train['tweet'].apply(lambda x:len([x for x in x.split() if x.startswith('#')]))
```

```python
train[['tweet','hashtags']].head()
```

```python
#Number of numerics
```

```python
train['numerics'] = train['tweet'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
```

```python
train[['tweet','numerics']].head()
```

```python
#Number of Uppercase words
```

```python
train['upper']=train['tweet'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
```

```python
train[['tweet','upper']].head()
```
